```python
#!/usr/bin/env python
# coding: utf-8

# # data description
# event index  
# Ti：乘客i按電梯按鈕的時間，∀i∈I  
# Ai：乘客i的起點樓層，∀i∈I  
# Di：乘客i的終點樓層，∀i∈I  

# # read file

# In[24]:


result = [[0 for j in range(9)] for i in range(10)]
for ir in range(10):
    for jr in range(9):
        
        filename = "data/test1-t%d-%d" % (60+jr*30, ir+1) 
        with open(filename+'.csv', newline='') as csvfile:
            rows = csv.reader(csvfile)
            i = 0
            data = []
            for row in rows:
                #if i % 2 == 0:
                    data.append(row)
                #i += 1


        # # preprocess

        # In[34]:


        index = list(map(int, data[0]))
        for i in range(len(index)):
            index[i] = i
        time = list(map(int, data[1]))
        arrival = list(map(int, data[2]))
        destination = list(map(int, data[3]))


        # # option

        # In[37]:


        num_of_passenger = 10 #number of passengers
        lowest_floor = 1 #lowest floor
        highest_floor = 10 #highest floor
        original_floor = getStart(filename+'.dat') #start at
        time_consumption = 5 #time consumption of transportation one floor


        # In[46]:


        passenger_status = [-1] * len(index) #-1 if passenger not arrived yet, 0  the passengers are waiting, 1 in the elevator, 2 done
        current_floor = original_floor
        current_time = 0
        max_waiting_so_far = -1
        elevator_status = 1 #1 upward, -1 downward
        last_time = max(time)
        mission_above = False
        mission_below = False
        integerize = 0
        while current_time <= last_time + 2 * time_consumption * (highest_floor - lowest_floor + 2):


            for passenger in index: #maintaining passenger waiting queue
                if current_time >= time[passenger] and passenger_status[passenger] == -1:
                    passenger_status[passenger] = 0


            for passenger in index:
                if passenger_status[passenger] == 0 and current_floor == arrival[passenger] and integerize==0:#接客
                    passenger_status[passenger] = 1
                if passenger_status[passenger] == 1 and current_floor == destination[passenger]and integerize==0:#送客
                    passenger_status[passenger] = 2
                    max_waiting_so_far = max(max_waiting_so_far, current_time - time[passenger])#更新最長等候時間

#             print("current floor:", current_floor)
            mission_above = False
            mission_below = False

            for passenger in index:
                if passenger_status[passenger] == 0:
                    if arrival[passenger] > current_floor:
                        mission_above = True
                    elif arrival[passenger] < current_floor:
                        mission_below = True
                elif passenger_status[passenger] == 1:
                    if destination[passenger] > current_floor:
                        mission_above = True
                    elif destination[passenger] < current_floor:
                        mission_below = True




            if elevator_status == 1:
                if mission_above:
                    integerize += 1
                    if integerize == time_consumption:
                        current_floor += 1
                        integerize = 0
                elif mission_below:
                    elevator_status = -1
                    integerize += 1
                    if integerize == time_consumption:
                        current_floor -= 1
                        integerize = 0
#                     print("current time:", current_time)
#                     print(passenger_status) 
#                     print("max waiting so far:", max_waiting_so_far, '\n\n')
                    current_time += 1
                    continue
                else:
                    elevator_status = 0
#                     print("current time:", current_time)
#                     print(passenger_status) 
#                     print("max waiting so far:", max_waiting_so_far, '\n\n')
                    current_time += 1
                    continue

            elif elevator_status == -1:
                if mission_below:
                    integerize += 1
                    if integerize == time_consumption:
                        current_floor -= 1
                        integerize = 0
                elif mission_above:
                    elevator_status = 1
                    current_floor += 1
#                     print("current time:", current_time)
#                     print(passenger_status) 
#                     print("max waiting so far:", max_waiting_so_far, '\n\n')
                    current_time += 1
                    continue
                else:
                    elevator_status = 0
#                     print("current time:", current_time)
#                     print(passenger_status) 
#                     print("max waiting so far:", max_waiting_so_far, '\n\n')
                    current_time += 1
                    continue

            else:
                if mission_above:
                    elevator_status = 1
                    integerize += 1
                    if integerize == time_consumption:
                        current_floor += 1
                        integerize = 0
                elif mission_below:
                    elevator_status = -1
                    integerize += 1
                    if integerize == time_consumption:
                        current_floor -= 1
                        integerize = 0


#             print("current time:", current_time)
#             print(passenger_status)
#             print("max waiting so far:", max_waiting_so_far, '\n\n')
            current_time += 1


#         print(max_waiting_so_far)
        result[ir][jr] = max_waiting_so_far
```

```python
def writeResult(filename, result, header):
    print(f'Write to{filename+".csv"}')
    with open(filename + '.csv', 'w') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        for i in range(len(result)):
            row = [str(i+1)]+list(map(str, result[i]))
            writer.writerow(row)
```
