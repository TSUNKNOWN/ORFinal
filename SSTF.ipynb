```python
import csv
import queue
import numpy as np
from  Utils import getStart
```

```python
result = [[j for j in range(9)] for i in range(10)]

for ir in range(10):
    for jr  in range(9):

        index, T, A, D = [],[],[],[] 
        speed = 5
        filename = f'data/test1-t{60+30*jr}-{ir+1}'
        with open(filename+'.csv', 'r') as f:
            rows = csv.reader(f)
            data = list(rows)
            for i in range(4):
                data[i] = list(map(int, data[i]))
            index, T, A, D, = np.array(data[0]),np.array(data[1]),np.array(data[2]),np.array(data[3])
        N = len(index)

        total_wait_time = 0
        max_wait_time = 0
        current_time = 0
        elevator_pos = getStart(filename+'.dat')
        arrivedNum = 0
        headto = elevator_pos
        onElevator = [0 for _ in range(N)] # wait:0, onElevator: 1, Arrived:2
        q = []
        head = 0
        tail = 0
        while arrivedNum < N:

            # Modify passenger status if it is fitted
            for i in range(N):
                if onElevator[i]==0 and A[i] == elevator_pos and T[i] <= current_time:
                    onElevator[i] = 1
                    q.append(i)
                    tail += 1
                if onElevator[i]==1 and D[i] == elevator_pos:
                    onElevator[i] = 2
                    wait_time = current_time - T[i]
                    total_wait_time += wait_time
                    max_wait_time = max(max_wait_time, wait_time)
                    arrivedNum += 1
            dellist = []
            for i in q:
                if onElevator[i] == 2:
                    dellist.append(i)
            for i in dellist:
                q.remove(i)

            # Decide elevator direction
            if len(q):
                headto = D[q[0]]
            else:
                for i in range(N):
                    if onElevator[i] == 0 and T[i] <= current_time:
                        headto = A[i]
                        break

            if headto > elevator_pos:
                elevator_pos += 1
                current_time += speed
            elif headto < elevator_pos:
                elevator_pos -= 1
                current_time += speed
            else:
                current_time += 1

        #     print(f'Time:{current_time}')
        #     print(q)
        #     print(onElevator)
        #     print(head, tail)
        print(f'Total wait time: {total_wait_time}')
        print(f'Max wait time:{max_wait_time}')
        print(f'Ending time: {current_time}')
        result[ir][jr] = max_wait_time
```

```python
def writeResult(filename, result, header):
    print(f'Write to{filename+".csv"}')
    with open(filename + '.csv', 'w') as f:
        writer = csv.writer(f)
        writer.writerow(header)
        for i in range(len(result)):
            row = [str(i+1)]+list(map(str, result[i]))
            writer.writerow(row)
```
